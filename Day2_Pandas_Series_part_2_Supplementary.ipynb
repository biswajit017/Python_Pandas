```python
# drop_duplicates

temp = pd.Series([1,1,2,2,3,3,4,4])
temp
```

drop_duplicates() in pandas is used to remove duplicate rows from a DataFrame or Series.

By default, it keeps the first occurrence and removes repeated values.

```python
temp.drop_duplicates()
```

It also gives additional flexibiity that is we can keep second repeated value and delete the first repeated value but bydefault it is first

```python
temp.drop_duplicates(keep='last')
```

Removes duplicate values/rows from temp and keeps the last occurrence of each duplicate.

```python
movies.drop_duplicates()
```

```python
temp.duplicated()
```

duplicated() in pandas is used to identify duplicate rows or values present in a series or not.

# dropna()

dropna() does not change the original data by default.

It creates and returns a new Series/DataFrame with null values removed, and stores the result there if assigned.

the datatype often becomes float.
Reason

Missing values (NaN) in pandas are stored as floating-point values.
So integer columns containing NaN are automatically converted to float.

```python
temp.dropna()
```

#fillna()
Suppose i dont want to drop those missing values then i can use fillna() function and replace those none value with some numbers or zero.

the datatype often becomes float.
Reason

Missing values (NaN) in pandas are stored as floating-point values.
So integer columns containing NaN are automatically converted to float.

```python
temp.fillna(0)
```

```python
temp.fillna(33)
```

If i want to replace those none value with mean of the dataset then i will write

```python
temp.fillna(temp.mean())
```

# copy()

copy() is used in Pandas to create a separate independent copy of a DataFrame or Series, so changes made to the copied object do not affect the original data.

```python
vk
```

```python
new = vk.head()

# wehen we are using head or tail() they will give us view of the data not copy or replica so whatever we will update it will directly update in the original dataset
```

```python
new
```

```python
new[1] = 100
```

head() may return a view of the original data. Therefore, modifying the returned object can sometimes update the original dataset.
on the above situation it is updated because prevously in vk, vk[1] was 1 after updated to 100 new one also changed as well as vk also changed
  so

To avoid this, use .copy().


```python
new = vk.head().copy()
```

```python
new[1] = 200
```
