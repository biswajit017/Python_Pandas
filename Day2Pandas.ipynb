```python
# write a function that can return the track record of 2 teams against each other


# Function to get head-to-head record
def track_record(team1, team2):

    # Filter matches between two teams
    record = ipl[
        ((ipl['Team1'] == team1) & (ipl['Team2'] == team2)) |
        ((ipl['Team1'] == team2) & (ipl['Team2'] == team1))
    ]

    # Count wins
    result = record['WinningTeam'].value_counts()

    return result

# Function call
print(track_record('Mumbai Indians', 'Chennai Super Kings'))
```
