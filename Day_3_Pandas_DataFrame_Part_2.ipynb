# value_counts(series and dataframe)

value_counts() in pandas is used to count the frequency of unique values in a Series or column. It returns how many times each value appears.

```python
marks = pd.DataFrame([
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,70,14],
    [80,70,14]
],columns=['iq','marks','package'])

marks
```

```python
marks.value_counts()
```

# sort_values(series and dataframe)

-> ascending -> na_position -> inplace -> multiple cols



sort_values() in pandas is used to sort the values of a Series or DataFrame in ascending or descending order.

```python
x = pd.Series([12,14,1,56,89])
x
```

```python
x.sort_values(ascending=False)
```

```python
movies = pd.read_csv('movies.csv')
movies.head(2)
```

```python
movies.sort_values('title_x',ascending=False)
```

```python
students = pd.DataFrame(
    {
        'name':['nitish','ankit','rupesh',np.nan,'mrityunjay',np.nan,'rishabh',np.nan,'aditya',np.nan],
        'college':['bit','iit','vit',np.nan,np.nan,'vlsi','ssit',np.nan,np.nan,'git'],
        'branch':['eee','it','cse',np.nan,'me','ce','civ','cse','bio',np.nan],
        'cgpa':[6.66,8.25,6.41,np.nan,5.6,9.0,7.4,10,7.4,np.nan],
        'package':[4,5,6,np.nan,6,7,8,9,np.nan,np.nan]

    }
)

students
```

```python
students.sort_values('name')
```

```python
students.sort_values('name',na_position='first',ascending=False)

# In default you will get last but if you want it in first then you need to write the attribute na_position='first'
```

```python
students
```

Fir Keeping it permanent we need to add another attribute

```python
students.sort_values('name',na_position='first',ascending=False,inplace=True)
```
