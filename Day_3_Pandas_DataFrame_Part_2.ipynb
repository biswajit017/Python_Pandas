# unique(series)

* unique() in pandas is used to get all unique/distinct values from a Series.

```python
temp.nunique() # nunique will not count the missing values
# Directly returns the number of unique values. no need to use len()
```

# dropna(series + dataframe) -> how parameter -> works like or

dropna() in pandas is used to remove missing (NaN) values from a Series or DataFrame.”

how parameter controls the condition:

* how='any' → removes row/column if ANY value is NaN (works like OR)
* how='all' → removes row/column only if ALL values are NaN

```python
students['name']
```

```python
students['name'].dropna()
```

```python
students
```

```python
students.dropna()
```

```python
students.dropna(how='any')
```

```python
students.dropna(how='all')
```

```python
students.dropna(subset=['name'])
```

```python
students.dropna(subset=['name','college'])
```

# fillna(series + dataframe)
fillna() in pandas is used to replace missing (NaN) values with a specified value or method.

```python
students['name'].fillna('unknown')
```

```python
students
```

```python
students.fillna(0) # column wises
```

```python
students['package'].fillna(students['package'].mean())
```

```python
students['name'].fillna(method='ffill')
# forward fill whatever the value will be there it will be forward to next missing value
```

```python
students['name'].fillna(method='bfill')
# This is the reverse of ffil and it will go backside or we can say upside of the missinng value
```
