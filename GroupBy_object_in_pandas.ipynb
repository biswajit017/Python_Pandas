### 📌 Important Points About `groupby()`

1. `groupby()` is used to divide data into groups based on common values present in one or more columns.

2. `groupby()` returns a GroupBy object, which means no result is produced until an operation is applied.

3. The grouped column automatically becomes the index in the output.

4. By default, `groupby()` sorts the group values in ascending order.

5. If you want to preserve the original order of the data, use `sort=False`.

6. `groupby()` follows the **Split → Apply → Combine** approach:

   * Split the data into groups.
   * Apply a function to each group.
   * Combine the results into a single output.

7. Aggregate functions such as `sum()`, `count()`, `mean()`, `min()`, and `max()` are commonly used with `groupby()`.

8. `groupby()` can be applied to both Series and DataFrame objects.

9. You can group data using a single column or multiple columns.

10. The `count()` function counts only non-null values within each group.

11. The `size()` function counts all rows in each group, including rows containing null values.

12. The `get_group()` method is used to retrieve all rows belonging to a specific group.

13. The `reset_index()` method can be used after grouping if you want the grouped column back as a normal column.

14. Grouping is most commonly performed on categorical columns such as City, Department, Gender, Team, etc.

15. The result of `groupby()` is usually smaller than the original dataset because multiple rows are summarized into groups.

# Find ranking of each movie in the group according to IMDB score


```python

def rank_movie(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
genres.apply(rank_movie)
```

# Find normalized IMDB rating group wise


```python

def normal(group):
  group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group

genres.apply(normal)
```

```python
ipl = pd.read_csv('/content/deliveries.csv')
ipl.head(1)
```

# Find the top 10 batsman in terms of runs


```python
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

# Find the batsman with max no of sixes


```python
six = ipl[ipl['batsman_runs'] == 6]


six.groupby('batsman')['batsman'].count().sort_values(ascending=False)
```

* ipl['batsman_runs'] == 6 checks every row and returns a boolean Series (True or False).
* True means the batsman scored a six, and False means he did not.
* When this boolean Series is placed inside ipl[...], Pandas treats it as a filter.
* Pandas keeps rows where the value is True and removes rows where it is False.
* As a result, ipl[ipl['batsman_runs'] == 6] returns only the deliveries where a six was hit.

# OR

```python
six.value_counts('batsman').sort_values(ascending=False).head(1).index[0]
```

# Find batsman with most number of 4's and 6's in last 5 overs


```python
temp_df = ipl[ipl['over'] > 15]
temp_df = temp_df[(temp_df['batsman_runs'] == 4 ) | (temp_df['batsman_runs'] == 6 )]
temp_df.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(10)
```

In Pandas, conditions return a Series of boolean values. Python's or operator works only with single boolean values, so we use | for element-wise OR operations on Series.

# Create a function that can return the highest score of any batsman


```python
temp_df = ipl[ipl['batsman'] == 'V Kohli']
temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
def highest(batsman):
  temp_df = ipl[ipl['batsman'] == batsman]
  return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```
